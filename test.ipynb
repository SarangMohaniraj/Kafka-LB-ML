```python
import json
import pandas as pd


# read images
drive.mount('/content/drive')
datadir = "/content/drive/My Drive/CS 511/Dataset/"


def json_to_df(file_path):
  temp = pd.DataFrame()
  c = 0
  with open(file_path, 'r', encoding='UTF-8') as file:
      while line := file.readline():
          c+=1
          if c>10000:
            break
          data = pd.read_json(line.rstrip(), lines=True)
          temp = pd.concat([temp, data], ignore_index = True)
  return temp

instance_usage_df = json_to_df(datadir+"instance_usage-000000000001.json")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import json


instance_usage_df['cpu_usage'] = instance_usage_df['average_usage'].apply(lambda x: x['cpus'])
instance_usage_df['memory_usage'] = instance_usage_df['average_usage'].apply(lambda x: x['memory'])
instance_usage_df['duration'] = (pd.to_datetime(instance_usage_df['end_time']) - pd.to_datetime(instance_usage_df['start_time'])).dt.total_seconds()

# Select relevant columns for training
X = instance_usage_df[['cpu_usage', 'memory_usage']]
y = instance_usage_df['duration']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05)

# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Print the coefficients of the linear regression model
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
